```python
import pandas as pd
import numpy as np

DATA = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'
```

```python
try:
    df_roubos = pd.read_csv(DATA, sep=';', encoding='latin-1')
    print(df_roubos['regiao'].unique())
    # print(df_roubos.head())

except Exception as e:
    print(f"Erro ao ler o arquivo CSV: {e}")
```

```python
try:
    # delimitando as variáveis
    df_roubos = df_roubos[['cisp', 'regiao', 'munic', 'roubo_transeunte']]

    df_correcao = df_roubos['regiao'].str.startswith('Grande Niter', na=False)
    # print(df_correcao)

    df_roubos.loc[df_correcao, 'regiao'] = 'Grande Niterói'
    print(df_roubos['regiao'].unique())

    # agrupando por cisp, regiao e munic
    df_roubos = df_roubos.groupby(['regiao', 'munic', 'cisp'], as_index=False)['roubo_transeunte'].sum()
    # display(df_roubos_grouped)

    print(df_roubos['roubo_transeunte'].describe())
            
except Exception as e:
    print(f"Erro ao limpar os dados: {e}")
```

```python
# processando os dados

try:
    array_roubos = np.array(df_roubos['roubo_transeunte'])

    media = np.mean(array_roubos)
    mediana = np.median(array_roubos)
    total = np.sum(array_roubos)

    # obter os quartis

    q1 = np.quantile(array_roubos, 0.25)
    q3 = np.quantile(array_roubos, 0.75)

    print('\nMedidas')
    print(f'\nMédia: {media}')
    print(f'Mediana: {mediana}')
    print(f'Total: {total}')
    print(f'Q1: {q1}')
    print(f'Q3: {q3}')



except Exception as e:
    print(f"Erro ao processar os dados: {e}")
```

```python
# criando uma coluna flag (mais e menos)
try:
    df_maiores = df_roubos[df_roubos['roubo_transeunte'] > q3].copy()
    df_maiores['flag'] = 'Maiores'
    
    df_menores = df_roubos[df_roubos['roubo_transeunte'] < q1].copy()
    df_menores['flag'] = 'Menores'

    # concatenando os dataframes
    df_roubos_flag = pd.concat([df_maiores, df_menores], ignore_index=True)

    display(df_roubos_flag)

except Exception as e:
     print(f"Erro ao processar os dados: {e}")
```

```python
# exportar como CSV e XLSX
try:

    df_roubos_flag.to_csv('roubo_transeunte.csv', index=False)
    df_roubos_flag.to_excel('roubo_transeunte.xlsx', index=False)
    print('Dataframe exportado com sucesso!')

except Exception as e:
    print(f'Erro ao exportar o arquivo: {e}')
```
